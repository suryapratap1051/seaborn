Seaborn:

Data Visualization Library
Importing the Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
     
```

Note : Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
     
```

```python

# visualize the data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter Plot

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',style='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
     
```

```python
titanic.head()
```

```python
titanic.shape
```

Count Plot

```python

sns.countplot(x='class',data=titanic)
```

```python

sns.countplot(x='survived',data=titanic)
```

Bar Chart



```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```
